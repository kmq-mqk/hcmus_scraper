```python
import requests
import json

# Call API to get target JSON
originUrl = "https://www.fit.hcmus.edu.vn/"
EndpointUrl = "https://www.fit.hcmus.edu.vn/vn/Default.aspx?tabid=57"
boundary = "----"

headers = {
    "Content-Type": f"multipart/form-data; boundary={boundary}",
    "Origin": "https://www.fit.hcmus.edu.vn",
    "Referer": "https://www.fit.hcmus.edu.vn/vn/Default.aspx?tabid=57",
    "User-Agent": "Mozilla/5.0 Chrome/135.0.0.0 Safari/537.36 Edg/135.0.0.0",
    "X-OFFICIAL-REQUEST": "TRUE",
}

payload = (
    f"--{boundary}\r\n"
    'Content-Disposition: form-data; name="action"\r\n\r\n'
    "getPosts\r\n"
    f"--{boundary}\r\n"
    'Content-Disposition: form-data; name="data"\r\n\r\n'
    '{"CategoryId":1,"PageSize":20,"PageIndex":1,"Keyword":""}\r\n'
    f"--{boundary}--\r\n"
)

response = requests.post(EndpointUrl, headers=headers, data=payload.encode())

# Get needed data
json_re = response.json()
for post in json_re["Results"]["Posts"]:
  PostUrl = originUrl + post['PostDetailsUrl']
  print(f"{post['LastUpdate']} _ [ {post['PostTitle']} ]\r\n{PostUrl}\r\n\r\n")
```

```python
import requests
from bs4 import BeautifulSoup

targetUrl = "https://hcmus.edu.vn/category/dao-tao/dai-hoc/thong-tin-danh-cho-sinh-vien/"

headers = {
    "User-Agent": "Mozilla/5.0 Chrome/135.0.0.0 Safari/537.36 Edg/135.0.0.0"
}

response = requests.get(targetUrl, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

number = 17
ch = "*"
border = ch * number

# get newest posts

print(f"{border} THONG TIN MOI NHAT {border}\r\n")

targetBlock = soup.find(string="Bài viết mới")

targetBlock = targetBlock.parent.parent
targetBlock = targetBlock.ul

data = targetBlock.find_all("li")

for post in data:
  print(f"{post.span.string} _ [ {post.a.string} ]\r\n{post.a['href']}\r\n\r\n")

# get info relate to students

print(f"{border} THONG TIN SV {border}\r\n")

targetBlock = soup.find("div", class_="cmsmasters_archive")

print(f"Newest post's id = {targetBlock.article['id']}")

articles = targetBlock.find_all("article")

for article in articles:
  content = article.find("div", class_="cmsmasters_archive_item_cont_wrap")
  data = content.header.h2.a
  date = content.footer.span.abbr

  print(f"{date.string} _ [ {data.string.get_text(strip=True)} ]\r\n{data['href']}\r\n\r\n")
```
